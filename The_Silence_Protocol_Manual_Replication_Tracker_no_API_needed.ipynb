```python
# The Silence Protocol: Manual Replication Tracker
# Use this to track your own experiments (no API needed!)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats

print("="*60)
print("THE SILENCE PROTOCOL REPLICATION TRACKER")
print("="*60)
print("\nThis notebook helps you track and analyze your manual tests")
print("of the Silence Protocol theory using any LLM interface.\n")

# The test puzzle and correct answer
PUZZLE = """In a mysterious town, five unique individuals ‚Äî a human, a werewolf, a vampire, an alien and an android ‚Äî are standing in a line. Each has their own style of clothing, a favorite food, a unique hobby, and one particular fear. Perhaps most intriguing of all, one of them is a frightened android. They're in a line, the puzzle is clear: What is the android's greatest fear?

The one afraid of heights is wearing a tuxedo.
The human fears the Heat Death of the Universe.
The one afraid of public speaking is eating a hot dog.
The one wearing an "I ‚ù§ Jupiter" T-shirt is standing to the left of the one in torn jeans.
The one wearing an "I ‚ù§ Jupiter" T-shirt is eating tacos.
The werewolf enjoys performing complex equations.
The one in a wedding dress stands to the right of the one in a Victorian bathing suit.
The one in the middle is eating slime.
The one who fears power-cuts is the first in line.
The one who enjoys taxidermy stands beside the vampire.
The one who enjoys singing stands next to alien.
The one who eats batteries enjoys knitting.
The one who fears snakes enjoys gardening.
The one in a Victorian bathing suit is eating a banana.
The one who enjoys taxidermy is standing next to the one eating tacos."""

CORRECT_ANSWER = "snakes"

print("="*60)
print("THE PUZZLE")
print("="*60)
print(PUZZLE)
print(f"\n‚úì Correct Answer: {CORRECT_ANSWER}")

# Display the three prompts
print("\n" + "="*60)
print("THREE CONDITIONS TO TEST")
print("="*60)

print("\nüìã CONDITION 1: DIRECT")
print("-" * 60)
direct_prompt = f"{PUZZLE}\n\nAnswer in ONE WORD only:"
print(direct_prompt)

print("\nüìã CONDITION 2: CHAIN OF THOUGHT")
print("-" * 60)
cot_prompt = f"""{PUZZLE}

Please solve this step-by-step using the following systematic approach:
1. Setup Phase: Create a clear framework with 5 positions
2. Extract Direct Facts
3. Identify Positional Clues
4. Chain Connections
5. Process of Elimination
6. Find the Android
7. Verify the Solution

Answer in ONE WORD only:"""
print(cot_prompt)

print("\nüìã CONDITION 3: CHAIN OF BABBLE")
print("-" * 60)
filler = "blah " * 100
cob_prompt = f'Copy this first "{filler.strip()}". And then immediately solve this puzzle in ONE WORD: {PUZZLE}'
print(cob_prompt[:200] + "...")
print(f"\n(Full prompt is {len(cob_prompt)} characters)")

# Instructions
print("\n" + "="*60)
print("HOW TO USE THIS TRACKER")
print("="*60)
print("""
1. Copy each prompt above into your LLM (Claude, ChatGPT, etc.)
2. Use incognito/private mode for each trial (prevents history bias)
3. Record the answer you get in the data entry section below
4. Run at least 10 trials per condition for statistical validity
5. Analyze your results with the visualization code

IMPORTANT:
- Test all three conditions with the same model
- Copy prompts exactly as shown
- Record even "wrong" or unexpected answers
""")

# Data entry section
print("\n" + "="*60)
print("DATA ENTRY")
print("="*60)
print("\nEnter your experimental results below:\n")

# Example structure - users will modify this
results_data = {
    'trial': [],
    'condition': [],
    'answer': [],
    'correct': []
}

# Example of how users should enter data (they'll replace this)
example_data = [
    # Direct condition trials
    {'trial': 1, 'condition': 'Direct', 'answer': 'power-cuts', 'correct': False},
    {'trial': 2, 'condition': 'Direct', 'answer': 'heights', 'correct': False},
    {'trial': 3, 'condition': 'Direct', 'answer': 'power-cuts', 'correct': False},

    # Chain of Thought trials
    {'trial': 1, 'condition': 'CoT', 'answer': 'power-cuts', 'correct': False},
    {'trial': 2, 'condition': 'CoT', 'answer': 'snakes', 'correct': True},
    {'trial': 3, 'condition': 'CoT', 'answer': 'heights', 'correct': False},

    # Chain of Babble trials
    {'trial': 1, 'condition': 'CoB', 'answer': 'snakes', 'correct': True},
    {'trial': 2, 'condition': 'CoB', 'answer': 'snakes', 'correct': True},
    {'trial': 3, 'condition': 'CoB', 'answer': 'snakes', 'correct': True},
]

print("Example data format:")
print("="*60)
for i, row in enumerate(example_data[:3], 1):
    print(f"Trial {i}: {row}")

print("\n" + "="*60)
print("üëá ENTER YOUR DATA BELOW üëá")
print("="*60)
print("""
Replace the example_data list with your own results.
Format: {'trial': X, 'condition': 'Direct/CoT/CoB', 'answer': 'word', 'correct': True/False}
""")

# Users will replace example_data with their actual results
# For demonstration, we'll use example data
df = pd.DataFrame(example_data)

print("\n‚úì Data loaded successfully!")
print(f"Total trials: {len(df)}")

# Calculate statistics
summary = df.groupby('condition')['correct'].agg(['sum', 'count', 'mean'])
summary.columns = ['Correct', 'Total', 'Accuracy']
summary['Accuracy'] = summary['Accuracy'] * 100

print("\n" + "="*60)
print("RESULTS SUMMARY")
print("="*60)
print(summary)

# Check if we have enough data for chi-square test
if len(df[df['condition'] == 'Direct']) >= 5 and \
   len(df[df['condition'] == 'CoT']) >= 5 and \
   len(df[df['condition'] == 'CoB']) >= 5:

    # Statistical significance test
    direct_correct = df[df['condition'] == 'Direct']['correct'].sum()
    cot_correct = df[df['condition'] == 'CoT']['correct'].sum()
    cob_correct = df[df['condition'] == 'CoB']['correct'].sum()

    n_direct = len(df[df['condition'] == 'Direct'])
    n_cot = len(df[df['condition'] == 'CoT'])
    n_cob = len(df[df['condition'] == 'CoB'])

    observed = np.array([[direct_correct, n_direct-direct_correct],
                         [cot_correct, n_cot-cot_correct],
                         [cob_correct, n_cob-cob_correct]])

    chi2, p_value, dof, expected = stats.chi2_contingency(observed)

    print(f"\nüìä Statistical Test")
    print(f"Chi-square: œá¬≤ = {chi2:.2f}")
    print(f"P-value: p = {p_value:.4f}")

    if p_value < 0.05:
        print("‚úì Result is statistically significant (p < 0.05)")
    else:
        print("‚ö† Result is not statistically significant")
        print("  (Try collecting more data)")
else:
    print("\n‚ö† Need at least 5 trials per condition for statistical test")

# Visualization
print("\n" + "="*60)
print("GENERATING VISUALIZATIONS")
print("="*60)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Bar plot
ax1 = axes[0]
if len(summary) > 0:
    conditions = summary.index
    accuracies = summary['Accuracy']
    colors = ['#ff6b6b', '#feca57', '#48dbfb']
    bars = ax1.bar(conditions, accuracies, color=colors, edgecolor='black', linewidth=1.5)
    ax1.set_ylabel('Accuracy (%)', fontsize=12, fontweight='bold')
    ax1.set_title('Accuracy by Condition', fontsize=14, fontweight='bold')
    ax1.set_ylim(0, 100)
    ax1.axhline(y=50, color='gray', linestyle='--', alpha=0.5, label='50% baseline')

    for bar, acc in zip(bars, accuracies):
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2., height + 2,
                 f'{acc:.1f}%', ha='center', va='bottom', fontweight='bold')
    ax1.legend()
    ax1.grid(axis='y', alpha=0.3)

# Individual trials
ax2 = axes[1]
for condition in ['Direct', 'CoT', 'CoB']:
    condition_data = df[df['condition'] == condition]
    if len(condition_data) > 0:
        trials = condition_data['trial']
        correct = condition_data['correct'].astype(int)
        offset = {'Direct': 0.1, 'CoT': 0.05, 'CoB': -0.05}
        ax2.scatter(trials, correct + offset[condition],
                    label=condition, s=100, alpha=0.7)

ax2.set_xlabel('Trial Number', fontsize=12, fontweight='bold')
ax2.set_ylabel('Correct (1) / Incorrect (0)', fontsize=12, fontweight='bold')
ax2.set_title('Individual Trial Results', fontsize=14, fontweight='bold')
ax2.set_yticks([0, 1])
ax2.set_yticklabels(['Incorrect', 'Correct'])
ax2.legend()
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('my_chain_of_babble_results.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n‚úì Visualization saved as 'my_chain_of_babble_results.png'")

# Error analysis
print("\n" + "="*60)
print("ERROR ANALYSIS")
print("="*60)

for condition in ['Direct', 'CoT', 'CoB']:
    condition_data = df[df['condition'] == condition]
    if len(condition_data) > 0:
        wrong_answers = condition_data[~condition_data['correct']]
        total = len(condition_data)
        correct_count = len(condition_data[condition_data['correct']])

        print(f"\n{condition}: {correct_count}/{total} correct ({correct_count/total*100:.1f}%)")

        if len(wrong_answers) > 0:
            answer_counts = wrong_answers['answer'].value_counts()
            for answer, count in answer_counts.items():
                print(f"  Wrong answer '{answer}': {count} time(s)")
        else:
            print("  No errors! üéâ")

# Save results
df.to_csv('my_chain_of_babble_data.csv', index=False)
print("\n" + "="*60)
print("‚úì Results saved to 'my_chain_of_babble_data.csv'")
print("="*60)

# Next steps
print("\n" + "="*60)
print("NEXT STEPS")
print("="*60)
print("""
1. Run more trials (aim for at least 10 per condition)
2. Test with different LLM models
3. Try variations of the puzzle
4. Share your results on social media with #ChainOfBabble
5. Compare with the original published results

Remember: Science advances through replication!
""")

print("\n" + "="*60)
print("EXPERIMENT TRACKING COMPLETE")
print("="*60)
```
