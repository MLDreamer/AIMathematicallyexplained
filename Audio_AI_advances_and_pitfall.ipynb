```python
"""
FROM BEETHOVEN TO VOICE-GOD: 3BLUE1BROWN-STYLE ANIMATION GENERATOR
===================================================================

This script generates slow-moving, pedagogical GIF animations that visualize
the evolution of audio AI from spectrograms to hyperbolic embeddings.

ANIMATIONS GENERATED:
1. SPECTROGRAM_FLATLAND.gif - How early AI flattened music
2. VECTOR_QUANTIZATION.gif - Continuous manifold â†’ discrete points
3. MULTIMODAL_ALIGNMENT.gif - Audio/text/image convergence
4. NEURAL_CODEC.gif - Audio â†” tokens transformation
5. PROSODIC_FEATURES.gif - Pitch/energy/rhythm in emotion
6. TEMPORAL_COMPRESSION.gif - Dynamic conversation â†’ static point
7. EUCLIDEAN_PRISON.gif - Flat geometry distorting distances
8. HYPERBOLIC_EMOTIONS.gif - Euclidean vs hyperbolic emotion space
9. GEODESIC_CLIFF.gif - The gap between now and future

STYLE: 3Blue1Brown aesthetic
- Dark background with vibrant colors
- Slow pacing (3-8 fps) for comprehension
- Clear annotations and mathematical overlays
- Educational text that builds progressively

Run this in Google Colab. All GIFs saved to /content/ for download.

Total generation time: ~20-25 minutes
"""

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from scipy.ndimage import gaussian_filter
from matplotlib.patches import Circle, FancyBboxPatch, FancyArrowPatch
from mpl_toolkits.mplot3d import proj3d
import warnings
warnings.filterwarnings('ignore')

# 3Blue1Brown color scheme
plt.style.use('dark_background')
BLUE = '#58C4DD'
YELLOW = '#FFFF00'
RED = '#FC6255'
GREEN = '#83C167'
PURPLE = '#CE6FFF'
ORANGE = '#FC9F5B'
WHITE = 'white'

print("=" * 80)
print("ðŸŽµ BEETHOVEN TO VOICE-GOD: ANIMATION GENERATOR")
print("=" * 80)
print("\nGenerating 9 pedagogical animations...")
print("Estimated time: ~20-25 minutes\n")

# ============================================================================
# ANIMATION 1: SPECTROGRAM_FLATLAND
# ============================================================================

print("ðŸŽ¬ Animation 1/9: SPECTROGRAM_FLATLAND.gif")
print("   Concept: How early AI reduced symphonies to heatmaps")
print("   Duration: 12 seconds")

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), facecolor='#0E1117')
for ax in [ax1, ax2]:
    ax.set_facecolor('#0E1117')

def animate_spectrogram(frame):
    ax1.clear()
    ax2.clear()
    for ax in [ax1, ax2]:
        ax.set_facecolor('#0E1117')

    # Create a waveform (Beethoven-esque)
    t = np.linspace(0, 4, 1000)
    # Simulate Moonlight Sonata opening (slow, melancholic arpeggios)
    wave = (np.sin(2*np.pi*1.5*t) * np.exp(-0.3*t) +  # Base note
            0.5*np.sin(2*np.pi*2.25*t) * np.exp(-0.25*t) +  # Third
            0.3*np.sin(2*np.pi*3*t) * np.exp(-0.2*t))  # Fifth

    # Add slow envelope
    envelope = 0.5 + 0.5*np.sin(0.5*t)
    wave = wave * envelope

    # Progress: gradually show more of the waveform
    show_up_to = int((frame / 80) * len(t))

    # Plot waveform
    ax1.plot(t[:show_up_to], wave[:show_up_to], color=BLUE, linewidth=2)
    ax1.set_xlim(0, 4)
    ax1.set_ylim(-2, 2)
    ax1.set_xlabel('Time (seconds)', fontsize=14, color=WHITE)
    ax1.set_ylabel('Amplitude', fontsize=14, color=WHITE)
    ax1.set_title('BEETHOVEN\'S MOONLIGHT SONATA (Simplified Waveform)\nContinuous, Flowing, Emotional',
                 fontsize=16, color=BLUE, weight='bold')
    ax1.grid(True, alpha=0.2, color=WHITE)

    # Create spectrogram (heatmap)
    # Simulate frequency bins
    freqs = np.linspace(50, 2000, 40)
    times = np.linspace(0, 4, 50)

    # Create spectrogram data (simplified)
    spec = np.zeros((len(freqs), len(times)))
    for i, f in enumerate(freqs):
        # Emphasize fundamental and harmonics
        if 100 < f < 200:  # Fundamental
            spec[i, :] = 0.8 + 0.2*np.sin(times*2)
        elif 200 < f < 350:  # Third
            spec[i, :] = 0.5 + 0.2*np.sin(times*1.5)
        elif 400 < f < 550:  # Fifth
            spec[i, :] = 0.3 + 0.1*np.sin(times)
        else:
            spec[i, :] = 0.05 * np.random.rand(len(times))

    # Fade in spectrogram as wave progresses
    fade = min(frame / 40, 1.0)
    spec_display = spec[:, :int(fade * len(times))]
    times_display = times[:int(fade * len(times))]

    if len(times_display) > 0:
        im = ax2.imshow(spec_display, aspect='auto', origin='lower',
                       extent=[0, times_display[-1], freqs[0], freqs[-1]],
                       cmap='hot', interpolation='bilinear')

        ax2.set_xlabel('Time (seconds)', fontsize=14, color=WHITE)
        ax2.set_ylabel('Frequency (Hz)', fontsize=14, color=WHITE)
        ax2.set_title('WHAT EARLY AI SAW: A Heatmap\n"Technically Accurate, Utterly Devoid of Soul"',
                     fontsize=16, color=RED, weight='bold')

        # Add annotation after spectrogram is fully shown
        if frame > 60:
            ax2.text(0.5, 0.95, 'Missing: Emotional arc, harmonic tension, temporal flow',
                    transform=ax2.transAxes, ha='center', va='top',
                    fontsize=12, color=YELLOW, weight='bold',
                    bbox=dict(boxstyle='round', facecolor='black', alpha=0.9))

    fig.suptitle('FROM MUSIC TO HEATMAP: The "Flat Earth" Fallacy',
                fontsize=20, color=YELLOW, weight='bold', y=0.98)

anim1 = FuncAnimation(fig, animate_spectrogram, frames=80, interval=150)
print("   Rendering...")
anim1.save('/content/SPECTROGRAM_FLATLAND.gif', writer=PillowWriter(fps=6.67))
plt.close()
print("   âœ“ Saved\n")

# ============================================================================
# ANIMATION 2: VECTOR_QUANTIZATION
# ============================================================================

print("ðŸŽ¬ Animation 2/9: VECTOR_QUANTIZATION.gif")
print("   Concept: Continuous manifold becomes discrete points")
print("   Duration: 14 seconds")

fig = plt.figure(figsize=(14, 10), facecolor='#0E1117')
ax = fig.add_subplot(111, projection='3d')
ax.set_facecolor('#0E1117')

def animate_quantization(frame):
    ax.clear()
    ax.set_facecolor('#0E1117')

    # Create a smooth manifold (continuous emotional space)
    u = np.linspace(0, 2*np.pi, 50)
    v = np.linspace(0, np.pi, 30)
    U, V = np.meshgrid(u, v)

    # Parametric surface (like a deformed sphere)
    X = (2 + 0.5*np.sin(3*U)) * np.sin(V)
    Y = (2 + 0.5*np.sin(3*U)) * np.cos(V)
    Z = 0.5*np.cos(3*U) + 0.3*np.sin(2*V)

    # Plot the continuous manifold
    progress = frame / 70

    if progress < 0.5:
        # Show smooth manifold
        alpha = min(progress * 2, 1.0)
        surf = ax.plot_surface(X, Y, Z, alpha=alpha*0.6, color=BLUE,
                              edgecolor='none', shade=True)
        stage = "CONTINUOUS MANIFOLD"
        stage_color = BLUE
        desc = "Infinite-dimensional space of expression"
    else:
        # Show quantized points
        # Sample discrete points from the manifold
        n_points = int((progress - 0.5) * 2 * 40)
        indices = np.random.choice(len(U.ravel()), min(n_points, len(U.ravel())), replace=False)

        X_flat = X.ravel()[indices]
        Y_flat = Y.ravel()[indices]
        Z_flat = Z.ravel()[indices]

        # Show fading manifold
        surf = ax.plot_surface(X, Y, Z, alpha=0.1, color=BLUE,
                              edgecolor='none', shade=True)

        # Show discrete points
        ax.scatter(X_flat, Y_flat, Z_flat, c=RED, s=50,
                  edgecolors=WHITE, linewidth=1, alpha=0.9)

        stage = "VECTOR QUANTIZATION"
        stage_color = RED
        desc = f"Reduced to {n_points} disconnected points"

    ax.set_xlabel('Dimension 1', fontsize=12, color=WHITE)
    ax.set_ylabel('Dimension 2', fontsize=12, color=WHITE)
    ax.set_zlabel('Dimension 3', fontsize=12, color=WHITE)
    ax.set_title(f'{stage}\n{desc}',
                fontsize=18, color=stage_color, weight='bold', pad=20)

    ax.view_init(elev=20, azim=frame*2)

    # Add formula
    if progress > 0.6:
        fig.text(0.5, 0.02, 'â„³continuous â†’ {vâ‚, vâ‚‚, ..., vâ‚™} âŠ‚ â„áµˆ',
                ha='center', fontsize=16, color=YELLOW,
                bbox=dict(boxstyle='round', facecolor='black', alpha=0.9))

anim2 = FuncAnimation(fig, animate_quantization, frames=70, interval=200)
print("   Rendering...")
anim2.save('/content/VECTOR_QUANTIZATION.gif', writer=PillowWriter(fps=5))
plt.close()
print("   âœ“ Saved\n")

# ============================================================================
# ANIMATION 3: MULTIMODAL_ALIGNMENT
# ============================================================================

print("ðŸŽ¬ Animation 3/9: MULTIMODAL_ALIGNMENT.gif")
print("   Concept: Audio/text/image converging in embedding space")
print("   Duration: 16 seconds")

fig = plt.figure(figsize=(14, 10), facecolor='#0E1117')
ax = fig.add_subplot(111, projection='3d')
ax.set_facecolor('#0E1117')

def animate_multimodal(frame):
    ax.clear()
    ax.set_facecolor('#0E1117')

    progress = frame / 80

    # Three modalities start far apart
    # Audio (red)
    audio_start = np.array([-3, 0, 0])
    # Text (blue)
    text_start = np.array([0, 3, 0])
    # Image (green)
    image_start = np.array([3, 0, 0])

    # They converge to center
    target = np.array([0, 0, 0])

    # Interpolate positions
    audio_pos = audio_start + progress * (target - audio_start)
    text_pos = text_start + progress * (target - text_start)
    image_pos = image_start + progress * (target - image_start)

    # Plot modality points
    ax.scatter(*audio_pos, s=500, c=RED, marker='o',
              edgecolors=WHITE, linewidth=3, label='Audio: "meow"')
    ax.scatter(*text_pos, s=500, c=BLUE, marker='s',
              edgecolors=WHITE, linewidth=3, label='Text: "cat"')
    ax.scatter(*image_pos, s=500, c=GREEN, marker='^',
              edgecolors=WHITE, linewidth=3, label='Image: ðŸ±')

    # Draw trails
    if frame > 10:
        trail_length = min(frame, 40)
        for i in range(1, trail_length, 2):
            t = (frame - i) / 80
            a_trail = audio_start + t * (target - audio_start)
            t_trail = text_start + t * (target - text_start)
            i_trail = image_start + t * (target - image_start)

            ax.plot([a_trail[0], a_trail[0]], [a_trail[1], a_trail[1]],
                   [a_trail[2], a_trail[2]], 'r.', alpha=0.3, markersize=2)
            ax.plot([t_trail[0], t_trail[0]], [t_trail[1], t_trail[1]],
                   [t_trail[2], t_trail[2]], 'b.', alpha=0.3, markersize=2)
            ax.plot([i_trail[0], i_trail[0]], [i_trail[1], i_trail[1]],
                   [i_trail[2], i_trail[2]], 'g.', alpha=0.3, markersize=2)

    # Show distances
    if progress > 0.3:
        audio_text_dist = np.linalg.norm(audio_pos - text_pos)
        ax.plot([audio_pos[0], text_pos[0]],
               [audio_pos[1], text_pos[1]],
               [audio_pos[2], text_pos[2]],
               color=YELLOW, linewidth=2, linestyle='--', alpha=0.6)

        mid = (audio_pos + text_pos) / 2
        ax.text(mid[0], mid[1], mid[2], f'd={audio_text_dist:.2f}',
               fontsize=12, color=YELLOW, weight='bold')

    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)
    ax.set_zlim(-4, 4)
    ax.set_xlabel('Dimension 1', fontsize=12, color=WHITE)
    ax.set_ylabel('Dimension 2', fontsize=12, color=WHITE)
    ax.set_zlabel('Dimension 3', fontsize=12, color=WHITE)

    if progress < 0.5:
        title = "BEFORE: Modalities Far Apart"
        color = RED
    else:
        title = "AFTER: Aligned in Shared Space"
        color = GREEN

    ax.set_title(f'{title}\nContrastive Learning Brings Concepts Together',
                fontsize=18, color=color, weight='bold', pad=20)

    ax.legend(loc='upper left', fontsize=11, framealpha=0.9)
    ax.view_init(elev=20, azim=45 + frame)

    # Add formula
    if progress > 0.7:
        fig.text(0.5, 0.02, 'Sim(A, T) = Î¦(A)Â·Î¨(T) / (||Î¦(A)|| ||Î¨(T)||)',
                ha='center', fontsize=14, color=YELLOW,
                bbox=dict(boxstyle='round', facecolor='black', alpha=0.9))

anim3 = FuncAnimation(fig, animate_multimodal, frames=80, interval=200)
print("   Rendering...")
anim3.save('/content/MULTIMODAL_ALIGNMENT.gif', writer=PillowWriter(fps=5))
plt.close()
print("   âœ“ Saved\n")

# ============================================================================
# ANIMATION 4: NEURAL_CODEC
# ============================================================================

print("ðŸŽ¬ Animation 4/9: NEURAL_CODEC.gif")
print("   Concept: Audio â†” tokens bidirectional transformation")
print("   Duration: 14 seconds")

fig, axes = plt.subplots(1, 3, figsize=(18, 6), facecolor='#0E1117')
for ax in axes:
    ax.set_facecolor('#0E1117')

def animate_codec(frame):
    for ax in axes:
        ax.clear()
        ax.set_facecolor('#0E1117')

    progress = frame / 70

    # LEFT: Continuous audio waveform
    t = np.linspace(0, 2, 200)
    wave = np.sin(2*np.pi*5*t) + 0.3*np.sin(2*np.pi*10*t)

    fade_in = min(progress * 2, 1.0)
    axes[0].plot(t[:int(fade_in*len(t))], wave[:int(fade_in*len(t))],
                color=BLUE, linewidth=2)
    axes[0].set_xlim(0, 2)
    axes[0].set_ylim(-2, 2)
    axes[0].set_xlabel('Time', fontsize=12, color=WHITE)
    axes[0].set_ylabel('Amplitude', fontsize=12, color=WHITE)
    axes[0].set_title('CONTINUOUS\nAUDIO', fontsize=14, color=BLUE, weight='bold')
    axes[0].grid(True, alpha=0.2, color=WHITE)

    # MIDDLE: Tokenization process
    if progress > 0.2:
        # Show discrete tokens appearing
        n_tokens = int((progress - 0.2) * 40)
        token_positions = np.linspace(0.1, 0.9, n_tokens)

        for i, pos in enumerate(token_positions):
            token_val = i % 256  # Simulate codebook index
            color_idx = token_val / 256
            color = cm.rainbow(color_idx)

            rect = FancyBboxPatch((pos-0.02, 0.3), 0.04, 0.4,
                                 boxstyle="round,pad=0.01",
                                 facecolor=color, edgecolor=WHITE,
                                 linewidth=2, alpha=0.9)
            axes[1].add_patch(rect)

            # Add token ID
            axes[1].text(pos, 0.5, f'{token_val}', ha='center', va='center',
                        fontsize=8, color='black', weight='bold')

        axes[1].set_xlim(0, 1)
        axes[1].set_ylim(0, 1)
        axes[1].axis('off')
        axes[1].set_title('DISCRETE\nTOKENS', fontsize=14, color=ORANGE, weight='bold')

        # Add arrow
        if progress > 0.3:
            axes[1].annotate('', xy=(0.5, 0.15), xytext=(0.5, 0.05),
                           arrowprops=dict(arrowstyle='->', color=YELLOW, lw=3))
            axes[1].text(0.5, 0.02, 'Quantization', ha='center',
                        fontsize=10, color=YELLOW)

    # RIGHT: Reconstructed audio
    if progress > 0.5:
        fade_reconstruct = min((progress - 0.5) * 2, 1.0)
        # Add slight quantization noise to show it's reconstructed
        wave_reconstructed = wave + 0.05 * np.random.randn(len(wave))

        axes[2].plot(t[:int(fade_reconstruct*len(t))],
                    wave_reconstructed[:int(fade_reconstruct*len(t))],
                    color=GREEN, linewidth=2, alpha=0.8)
        axes[2].set_xlim(0, 2)
        axes[2].set_ylim(-2, 2)
        axes[2].set_xlabel('Time', fontsize=12, color=WHITE)
        axes[2].set_ylabel('Amplitude', fontsize=12, color=WHITE)
        axes[2].set_title('RECONSTRUCTED\nAUDIO', fontsize=14, color=GREEN, weight='bold')
        axes[2].grid(True, alpha=0.2, color=WHITE)

        # Show quality metric
        if progress > 0.8:
            axes[2].text(0.5, 0.95, 'Perceptually\nLossless',
                        transform=axes[2].transAxes, ha='center', va='top',
                        fontsize=11, color=GREEN, weight='bold',
                        bbox=dict(boxstyle='round', facecolor='black', alpha=0.9))

    fig.suptitle('NEURAL CODEC: Continuous â†” Discrete\n"The LLM Can Finally Hear and Speak Directly"',
                fontsize=18, color=YELLOW, weight='bold', y=0.98)

anim4 = FuncAnimation(fig, animate_codec, frames=70, interval=200)
print("   Rendering...")
anim4.save('/content/NEURAL_CODEC.gif', writer=PillowWriter(fps=5))
plt.close()
print("   âœ“ Saved\n")

# ============================================================================
# ANIMATION 5: PROSODIC_FEATURES
# ============================================================================

print("ðŸŽ¬ Animation 5/9: PROSODIC_FEATURES.gif")
print("   Concept: How pitch, energy, rhythm encode emotion")
print("   Duration: 16 seconds")

fig, axes = plt.subplots(4, 1, figsize=(14, 12), facecolor='#0E1117')
for ax in axes:
    ax.set_facecolor('#0E1117')

def animate_prosody(frame):
    for ax in axes:
        ax.clear()
        ax.set_facecolor('#0E1117')

    t = np.linspace(0, 5, 500)
    progress = frame / 80
    show_up_to = int(progress * len(t))

    # Different emotions have different prosodic patterns
    if progress < 0.33:
        # Happy: High pitch, high energy, fast rhythm
        emotion = "HAPPY"
        color = YELLOW
        pitch = 200 + 50*np.sin(2*np.pi*2*t)  # Higher baseline, more variation
        energy = 0.8 + 0.2*np.sin(2*np.pi*1.5*t)  # High energy
        rhythm_gaps = [0.3, 0.6, 0.9, 1.2]  # Fast
    elif progress < 0.66:
        # Sad: Low pitch, low energy, slow rhythm
        emotion = "SAD"
        color = BLUE
        pitch = 120 + 20*np.sin(2*np.pi*0.5*t)  # Lower baseline, less variation
        energy = 0.3 + 0.1*np.sin(2*np.pi*0.8*t)  # Low energy
        rhythm_gaps = [0.5, 1.2, 2.0, 2.9]  # Slow
    else:
        # Angry: Variable pitch, high energy, irregular rhythm
        emotion = "ANGRY"
        color = RED
        pitch = 150 + 80*np.sin(2*np.pi*3*t) * np.exp(-0.3*t)  # Sharp variations
        energy = 0.9 + 0.1*np.sin(2*np.pi*4*t)  # Very high energy
        rhythm_gaps = [0.2, 0.45, 0.6, 0.95]  # Irregular

    # Plot 1: Waveform
    wave = np.sin(2*np.pi*pitch[:show_up_to]/40) * energy[:show_up_to]
    axes[0].plot(t[:show_up_to], wave, color=color, linewidth=1.5)
    axes[0].set_xlim(0, 5)
    axes[0].set_ylim(-1, 1)
    axes[0].set_ylabel('Waveform', fontsize=11, color=WHITE)
    axes[0].set_title(f'EMOTION: {emotion}', fontsize=16, color=color, weight='bold')
    axes[0].grid(True, alpha=0.2, color=WHITE)

    # Plot 2: Pitch contour
    axes[1].plot(t[:show_up_to], pitch[:show_up_to], color=PURPLE, linewidth=2)
    axes[1].set_xlim(0, 5)
    axes[1].set_ylim(80, 280)
    axes[1].set_ylabel('Pitch fâ‚€ (Hz)', fontsize=11, color=WHITE)
    axes[1].grid(True, alpha=0.2, color=WHITE)
    axes[1].axhline(y=150, color=WHITE, linestyle='--', alpha=0.3)

    # Plot 3: Energy
    axes[2].plot(t[:show_up_to], energy[:show_up_to], color=ORANGE, linewidth=2)
    axes[2].fill_between(t[:show_up_to], 0, energy[:show_up_to],
                         color=ORANGE, alpha=0.3)
    axes[2].set_xlim(0, 5)
    axes[2].set_ylim(0, 1.2)
    axes[2].set_ylabel('Energy', fontsize=11, color=WHITE)
    axes[2].grid(True, alpha=0.2, color=WHITE)

    # Plot 4: Rhythm (syllable timing)
    axes[3].set_xlim(0, 5)
    axes[3].set_ylim(0, 1)
    axes[3].set_ylabel('Rhythm', fontsize=11, color=WHITE)
    axes[3].set_xlabel('Time (seconds)', fontsize=11, color=WHITE)

    # Show syllable markers that have passed
    for gap in rhythm_gaps:
        if gap < t[show_up_to] if show_up_to > 0 else 0:
            axes[3].axvline(x=gap, color=GREEN, linewidth=3, alpha=0.8)
            axes[3].scatter([gap], [0.5], s=200, c=GREEN, marker='|',
                          linewidths=5, zorder=10)

    axes[3].grid(True, alpha=0.2, color=WHITE)

    fig.suptitle('PROSODIC FEATURES ENCODE EMOTION\n"AI Can Now Hear How You Say It, Not Just What You Say"',
                fontsize=18, color=YELLOW, weight='bold', y=0.98)

anim5 = FuncAnimation(fig, animate_prosody, frames=80, interval=200)
print("   Rendering...")
anim5.save('/content/PROSODIC_FEATURES.gif', writer=PillowWriter(fps=5))
plt.close()
print("   âœ“ Saved\n")

# ============================================================================
# ANIMATION 6: TEMPORAL_COMPRESSION
# ============================================================================

print("ðŸŽ¬ Animation 6/9: TEMPORAL_COMPRESSION.gif")
print("   Concept: Dynamic conversation collapses to single point")
print("   Duration: 14 seconds")

fig = plt.figure(figsize=(14, 10), facecolor='#0E1117')
ax = fig.add_subplot(111, projection='3d')
ax.set_facecolor('#0E1117')

def animate_compression(frame):
    ax.clear()
    ax.set_facecolor('#0E1117')

    progress = frame / 70

    # Create a trajectory through emotional space
    t_param = np.linspace(0, 4*np.pi, 100)
    x_traj = np.sin(t_param) * (1 + 0.3*np.sin(3*t_param))
    y_traj = np.cos(t_param) * (1 + 0.3*np.sin(3*t_param))
    z_traj = 0.5*t_param/np.pi - 1

    # Show trajectory building up
    show_points = int(progress * len(t_param))

    if show_points > 0:
        # Draw the trajectory
        ax.plot(x_traj[:show_points], y_traj[:show_points], z_traj[:show_points],
               color=BLUE, linewidth=3, alpha=0.8)

        # Current position
        ax.scatter([x_traj[show_points-1]], [y_traj[show_points-1]], [z_traj[show_points-1]],
                  s=300, c=YELLOW, marker='o', edgecolors=WHITE, linewidth=3)

        # Add temporal markers
        for i in range(0, show_points, 15):
            ax.scatter([x_traj[i]], [y_traj[i]], [z_traj[i]],
                      s=50, c=WHITE, alpha=0.5)

    # After trajectory is complete, show compression
    if progress > 0.7:
        # Compute centroid (average position)
        centroid = np.array([np.mean(x_traj[:show_points]),
                           np.mean(y_traj[:show_points]),
                           np.mean(z_traj[:show_points])])

        # Fade trajectory, emphasize centroid
        fade = (progress - 0.7) / 0.3

        ax.plot(x_traj[:show_points], y_traj[:show_points], z_traj[:show_points],
               color=BLUE, linewidth=3, alpha=0.8*(1-fade))

        ax.scatter([centroid[0]], [centroid[1]], [centroid[2]],
                  s=500*(1+fade), c=RED, marker='*',
                  edgecolors=WHITE, linewidth=3, zorder=10)

        # Show compression arrows
        if fade > 0.5:
            for i in [0, 25, 50, 75]:
                ax.plot([x_traj[i], centroid[0]],
                       [y_traj[i], centroid[1]],
                       [z_traj[i], centroid[2]],
                       color=RED, linewidth=2, linestyle='--', alpha=0.6)

    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)
    ax.set_zlim(-2, 2)
    ax.set_xlabel('Emotional Dim 1', fontsize=12, color=WHITE)
    ax.set_ylabel('Emotional Dim 2', fontsize=12, color=WHITE)
    ax.set_zlabel('Emotional Dim 3', fontsize=12, color=WHITE)

    if progress < 0.7:
        title = "CONVERSATION AS TRAJECTORY\nÎ³(t): [0,T] â†’ â„³"
        color = BLUE
    else:
        title = "COMPRESSED TO SINGLE VECTOR\nx(t) âˆˆ â„áµ€ â†’ v âˆˆ â„áµˆ"
        color = RED

    ax.set_title(title, fontsize=18, color=color, weight='bold', pad=20)
    ax.view_init(elev=20, azim=45 + frame*2)

    if progress > 0.8:
        fig.text(0.5, 0.02, '"Like reading only the last sentence of a novel"',
                ha='center', fontsize=14, color=YELLOW, style='italic',
                bbox=dict(boxstyle='round', facecolor='black', alpha=0.9))

anim6 = FuncAnimation(fig, animate_compression, frames=70, interval=200)
print("   Rendering...")
anim6.save('/content/TEMPORAL_COMPRESSION.gif', writer=PillowWriter(fps=5))
plt.close()
print("   âœ“ Saved\n")

# ============================================================================
# ANIMATION 7: EUCLIDEAN_PRISON
# ============================================================================

print("ðŸŽ¬ Animation 7/9: EUCLIDEAN_PRISON.gif")
print("   Concept: How flat geometry distorts emotional distances")
print("   Duration: 14 seconds")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7), facecolor='#0E1117')
for ax in [ax1, ax2]:
    ax.set_facecolor('#0E1117')

def animate_euclidean(frame):
    ax1.clear()
    ax2.clear()
    for ax in [ax1, ax2]:
        ax.set_facecolor('#0E1117')

    # Define emotional states
    emotions = {
        'happy': (0, 2),
        'elated': (0, 3.5),  # Intensification of happy
        'angry': (3, 0),  # Different category
        'serene': (0, 1.5),  # Subtle variant of happy
    }

    progress = frame / 70

    # LEFT: Show Euclidean distances
    for emotion, pos in emotions.items():
        color = YELLOW if emotion == 'happy' else BLUE if 'e' in emotion else RED if emotion == 'angry' else GREEN
        ax1.scatter(*pos, s=500, c=color, edgecolors=WHITE, linewidth=3, zorder=10)
        ax1.text(pos[0], pos[1] + 0.3, emotion.upper(), ha='center',
                fontsize=12, color=color, weight='bold')

    # Show distances progressively
    if progress > 0.2:
        # happy â†’ elated
        ax1.plot([emotions['happy'][0], emotions['elated'][0]],
                [emotions['happy'][1], emotions['elated'][1]],
                color=BLUE, linewidth=3, linestyle='--')
        dist_he = np.linalg.norm(np.array(emotions['happy']) - np.array(emotions['elated']))
        mid_he = ((np.array(emotions['happy']) + np.array(emotions['elated'])) / 2)
        ax1.text(mid_he[0] + 0.5, mid_he[1], f'd={dist_he:.2f}',
                fontsize=11, color=BLUE, weight='bold')

    if progress > 0.5:
        # happy â†’ angry
        ax1.plot([emotions['happy'][0], emotions['angry'][0]],
                [emotions['happy'][1], emotions['angry'][1]],
                color=RED, linewidth=3, linestyle='--')
        dist_ha = np.linalg.norm(np.array(emotions['happy']) - np.array(emotions['angry']))
        mid_ha = ((np.array(emotions['happy']) + np.array(emotions['angry'])) / 2)
        ax1.text(mid_ha[0], mid_ha[1] - 0.3, f'd={dist_ha:.2f}',
                fontsize=11, color=RED, weight='bold')

    if progress > 0.7:
        # Highlight the problem
        ax1.text(0.5, 0.95, f'PROBLEM: Similar distances!\nIntensification â‰ˆ Category change',
                transform=ax1.transAxes, ha='center', va='top',
                fontsize=13, color=YELLOW, weight='bold',
                bbox=dict(boxstyle='round', facecolor='black', edgecolor=RED, linewidth=3, alpha=0.9))

    ax1.set_xlim(-1, 4)
    ax1.set_ylim(-1, 4.5)
    ax1.set_xlabel('Dimension 1', fontsize=12, color=WHITE)
    ax1.set_ylabel('Dimension 2', fontsize=12, color=WHITE)
    ax1.set_title('EUCLIDEAN SPACE\n"Flat Geometry"',
                 fontsize=16, color=RED, weight='bold')
    ax1.grid(True, alpha=0.2, color=WHITE)
    ax1.set_aspect('equal')

    # RIGHT: Show what we SHOULD have (hierarchical representation)
    ax2.text(0.5, 0.5, 'Coming in GIF 8:\nHYPERBOLIC SPACE\nsolves this!',
            transform=ax2.transAxes, ha='center', va='center',
            fontsize=20, color=GREEN, weight='bold',
            bbox=dict(boxstyle='round', facecolor='black', alpha=0.9))
    ax2.axis('off')

    fig.suptitle('THE EUCLIDEAN PRISON\n"Why Cosine Similarity Fails for Emotions"',
                fontsize=20, color=YELLOW, weight='bold', y=0.98)

anim7 = FuncAnimation(fig, animate_euclidean, frames=70, interval=200)
print("   Rendering...")
anim7.save('/content/EUCLIDEAN_PRISON.gif', writer=PillowWriter(fps=5))
plt.close()
print("   âœ“ Saved\n")

# ============================================================================
# ANIMATION 8: HYPERBOLIC_EMOTIONS
# ============================================================================

print("ðŸŽ¬ Animation 8/9: HYPERBOLIC_EMOTIONS.gif")
print("   Concept: Compare Euclidean vs hyperbolic emotion spaces")
print("   Duration: 18 seconds")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8), facecolor='#0E1117', subplot_kw=dict(projection='polar'))
for ax in [ax1, ax2]:
    ax.set_facecolor('#0E1117')

def animate_hyperbolic(frame):
    ax1.clear()
    ax2.clear()
    for ax in [ax1, ax2]:
        ax.set_facecolor('#0E1117')

    progress = frame / 90

    # Emotion categories and hierarchies
    positive_emotions = ['joy', 'elated', 'content', 'serene']
    negative_emotions = ['sad', 'angry', 'anxious', 'frustrated']

    # LEFT: Euclidean (polar plot, but distances are uniform)
    theta_pos = np.linspace(0, np.pi/2, len(positive_emotions))
    theta_neg = np.linspace(np.pi, 3*np.pi/2, len(negative_emotions))

    r_uniform = 1.0

    fade = min(progress * 2, 1.0)

    for i, emotion in enumerate(positive_emotions):
        ax1.scatter(theta_pos[i], r_uniform, s=300*fade, c=YELLOW,
                   edgecolors=WHITE, linewidth=2, zorder=10)
        if fade > 0.5:
            ax1.text(theta_pos[i], r_uniform + 0.15, emotion,
                    ha='center', fontsize=9, color=YELLOW)

    for i, emotion in enumerate(negative_emotions):
        ax1.scatter(theta_neg[i], r_uniform, s=300*fade, c=BLUE,
                   edgecolors=WHITE, linewidth=2, zorder=10)
        if fade > 0.5:
            ax1.text(theta_neg[i], r_uniform + 0.15, emotion,
                    ha='center', fontsize=9, color=BLUE)

    ax1.set_ylim(0, 1.5)
    ax1.set_title('EUCLIDEAN\nAll emotions equally distant',
                 fontsize=14, color=RED, weight='bold', pad=20)
    ax1.grid(True, alpha=0.2)

    # RIGHT: Hyperbolic (distances grow exponentially with hierarchy)
    if progress > 0.3:
        fade2 = min((progress - 0.3) / 0.7, 1.0)

        # In hyperbolic space, similar emotions cluster tightly
        # Different categories are exponentially farther
        r_subcategory = 0.5
        r_category = 1.2  # Much farther due to hyperbolic metric

        # Positive branch
        for i, emotion in enumerate(positive_emotions):
            ax2.scatter(theta_pos[i], r_subcategory, s=300*fade2, c=YELLOW,
                       edgecolors=WHITE, linewidth=2, zorder=10)
            if fade2 > 0.5:
                ax2.text(theta_pos[i], r_subcategory + 0.1, emotion,
                        ha='center', fontsize=9, color=YELLOW)

        # Negative branch
        for i, emotion in enumerate(negative_emotions):
            ax2.scatter(theta_neg[i], r_subcategory, s=300*fade2, c=BLUE,
                       edgecolors=WHITE, linewidth=2, zorder=10)
            if fade2 > 0.5:
                ax2.text(theta_neg[i], r_subcategory + 0.1, emotion,
                        ha='center', fontsize=9, color=BLUE)

        # Show hierarchical clustering
        if fade2 > 0.6:
            # Draw branches
            theta_pos_center = np.mean(theta_pos)
            theta_neg_center = np.mean(theta_neg)

            ax2.plot([theta_pos_center, theta_pos_center], [0, r_subcategory],
                    color=YELLOW, linewidth=3, alpha=0.6)
            ax2.plot([theta_neg_center, theta_neg_center], [0, r_subcategory],
                    color=BLUE, linewidth=3, alpha=0.6)

            # Root
            ax2.scatter(0, 0, s=500, c=WHITE, marker='o',
                       edgecolors=PURPLE, linewidth=4, zorder=15)
            ax2.text(0, 0.05, 'NEUTRAL',
                    ha='center', fontsize=10, color=WHITE, weight='bold')

        ax2.set_ylim(0, 1.5)
        ax2.set_title('HYPERBOLIC\nHierarchical clustering preserved',
                     fontsize=14, color=GREEN, weight='bold', pad=20)
        ax2.grid(True, alpha=0.2)

    fig.suptitle('EUCLIDEAN vs HYPERBOLIC EMOTION SPACE\n"Negative Curvature Captures Hierarchy"',
                fontsize=20, color=PURPLE, weight='bold', y=0.98)

anim8 = FuncAnimation(fig, animate_hyperbolic, frames=90, interval=200)
print("   Rendering...")
anim8.save('/content/HYPERBOLIC_EMOTIONS.gif', writer=PillowWriter(fps=5))
plt.close()
print("   âœ“ Saved\n")

# ============================================================================
# ANIMATION 9: GEODESIC_CLIFF
# ============================================================================

print("ðŸŽ¬ Animation 9/9: GEODESIC_CLIFF.gif")
print("   Concept: The gap between current AI and true understanding")
print("   Duration: 16 seconds")

fig = plt.figure(figsize=(14, 10), facecolor='#0E1117')
ax = fig.add_subplot(111, projection='3d')
ax.set_facecolor('#0E1117')

def animate_cliff(frame):
    ax.clear()
    ax.set_facecolor('#0E1117')

    progress = frame / 80

    # Create two levels: current AI (flat) and future AI (curved)
    x = np.linspace(-3, 3, 50)
    y = np.linspace(-3, 3, 50)
    X, Y = np.meshgrid(x, y)

    # Current AI: flat surface at z=0
    Z_current = np.zeros_like(X)

    # Future AI: curved hyperbolic surface
    Z_future = 2 - 0.5 * (X**2 + Y**2)

    # Plot current AI surface
    fade_current = min(progress * 2, 1.0)
    surf1 = ax.plot_surface(X, Y, Z_current, alpha=fade_current*0.4,
                           color=RED, edgecolor='none', shade=True)

    # Add AI model markers on current surface
    if progress > 0.2:
        models = [
            ('2018\nMFCC', -2, -2),
            ('2020\nWav2Vec', 0, -2),
            ('2023\nCLAP', 2, 0),
            ('2026\nGPT-4o', 0, 2)
        ]

        show_models = int((progress - 0.2) * len(models) / 0.3)
        for i in range(min(show_models, len(models))):
            model_name, mx, my = models[i]
            ax.scatter([mx], [my], [0], s=300, c=WHITE,
                      edgecolors=RED, linewidth=3, zorder=10)
            ax.text(mx, my, 0.3, model_name, ha='center',
                   fontsize=9, color=WHITE, weight='bold')

    # Plot future AI surface (appears later)
    if progress > 0.5:
        fade_future = (progress - 0.5) / 0.5
        surf2 = ax.plot_surface(X, Y, Z_future, alpha=fade_future*0.6,
                               color=BLUE, edgecolor='none', shade=True)

        # Add aspiration markers
        if progress > 0.7:
            aspirations = [
                ('Temporal\nTrajectories', -1.5, 0),
                ('Hyperbolic\nEmbeddings', 1.5, 1),
                ('Ricci Flow\nDynamics', 0, -1.5)
            ]

            for asp_name, ax_pos, ay_pos in aspirations:
                az_pos = 2 - 0.5 * (ax_pos**2 + ay_pos**2)
                ax.scatter([ax_pos], [ay_pos], [az_pos], s=300, c=GREEN,
                          marker='*', edgecolors=WHITE, linewidth=2, zorder=10)
                ax.text(ax_pos, ay_pos, az_pos + 0.4, asp_name, ha='center',
                       fontsize=9, color=GREEN, weight='bold')

        # Draw "the cliff" - the gap
        if progress > 0.8:
            # Vertical lines showing the gap
            for gx, gy in [(-2, 0), (0, 2), (2, -1)]:
                gz_future = 2 - 0.5 * (gx**2 + gy**2)
                ax.plot([gx, gx], [gy, gy], [0, gz_future],
                       color=YELLOW, linewidth=3, linestyle='--', alpha=0.8)

            # "THE CLIFF" label
            ax.text(0, 0, 1, 'THE\nGEODESIC\nCLIFF', ha='center', va='center',
                   fontsize=16, color=YELLOW, weight='bold',
                   bbox=dict(boxstyle='round', facecolor='black',
                           edgecolor=YELLOW, linewidth=3, alpha=0.95))

    ax.set_xlim(-3, 3)
    ax.set_ylim(-3, 3)
    ax.set_zlim(-1, 3)
    ax.set_xlabel('Semantic Dimension', fontsize=11, color=WHITE)
    ax.set_ylabel('Emotional Dimension', fontsize=11, color=WHITE)
    ax.set_zlabel('Understanding Depth', fontsize=11, color=WHITE)

    if progress < 0.5:
        title = "CURRENT AI: Flat Embeddings"
        color = RED
    elif progress < 0.7:
        title = "FUTURE AI: Curved Manifolds"
        color = BLUE
    else:
        title = "THE GEODESIC CLIFF"
        color = YELLOW

    ax.set_title(f'{title}\n"Where Mathematical Models Meet Their Limits"',
                fontsize=18, color=color, weight='bold', pad=20)

    ax.view_init(elev=25, azim=45 + frame*1.5)

    if progress > 0.9:
        fig.text(0.5, 0.02, '"How do we cross from flat to curved understanding?"',
                ha='center', fontsize=13, color=YELLOW, style='italic',
                bbox=dict(boxstyle='round', facecolor='black', alpha=0.9))

anim9 = FuncAnimation(fig, animate_cliff, frames=80, interval=200)
print("   Rendering...")
anim9.save('/content/GEODESIC_CLIFF.gif', writer=PillowWriter(fps=5))
plt.close()
print("   âœ“ Saved\n")

# ============================================================================
# COMPLETION
# ============================================================================

print("=" * 80)
print("âœ¨ ALL ANIMATIONS GENERATED SUCCESSFULLY!")
print("=" * 80)
print("\nGenerated files (downloadable from Files panel):")
print("  1. /content/SPECTROGRAM_FLATLAND.gif")
print("  2. /content/VECTOR_QUANTIZATION.gif")
print("  3. /content/MULTIMODAL_ALIGNMENT.gif")
print("  4. /content/NEURAL_CODEC.gif")
print("  5. /content/PROSODIC_FEATURES.gif")
print("  6. /content/TEMPORAL_COMPRESSION.gif")
print("  7. /content/EUCLIDEAN_PRISON.gif")
print("  8. /content/HYPERBOLIC_EMOTIONS.gif")
print("  9. /content/GEODESIC_CLIFF.gif")
print("\nðŸ“¥ To download:")
print("  â€¢ Click folder icon â†’ /content/ â†’ Right-click each GIF â†’ Download")
print("\nðŸŽ¨ Animation characteristics:")
print("  â€¢ 3Blue1Brown dark aesthetic")
print("  â€¢ Slow pacing (5-6.67 fps)")
print("  â€¢ Progressive concept building")
print("  â€¢ Mathematical overlays")
print("  â€¢ Self-explanatory narratives")
print("\nðŸ’¡ Usage:")
print("  Replace [GIF X: FILENAME.gif] in the Medium article with embeds")
print("\nðŸŽµ From Beethoven to Voice-God: The journey is complete.")
print("=" * 80)
```
